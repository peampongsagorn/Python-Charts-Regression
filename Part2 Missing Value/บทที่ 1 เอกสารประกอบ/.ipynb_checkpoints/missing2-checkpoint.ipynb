# Missing value 

## 1. Mark ค่า NaN ด้วยคอลัมน์

```python
import pandas as pd
df = pd.read_excel('dataset.xlsx', sheet_name='missing')
```

```python
df.info()
```

```python
df_add_col = df.copy()
df_add_col['mark'] = df_add_col['Height'].isnull()
```
