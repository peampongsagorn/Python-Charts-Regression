# Linear regression

### 1. Import and visualize data

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
USAhousing = pd.read_csv('USA_Housing.csv')
```

```python
USAhousing.columns
```

```python
sns.heatmap(USAhousing.corr(), annot=True) # corrความสัมพันธ์ของแต่ละตัวแปร ใช้heat mapดูว่าข้อมูลตัวไหนมีควาามเกี่ยวข้องเป็นผลกันมาก
```

### 2. Split data set to training data and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
USAhousing.info()
```

```python
x=USAhousing.iloc[:,0:4]
```

```python
y=USAhousing['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
```

### 3. Build the linear regressing model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train) #ทำการ train
```

```python
lm.intercept_ #จุดตัดของเส้นตรงกับแกน y ที่ x = 0
```

```python
lm.coef_
```

```python
coeff_df = pd.DataFrame(lm.coef_,x.columns,columns=['Coefficient'])
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
coeff_df
```

### 4. Compare between original data and the result of prediction from the model

```python
predictions = lm.predict(X_test)
```

```python
predictions
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,predictions)
```

```python
y_test == predictions
```

```python
sns.distplot((y_test - predictions),kde=False,bins=10);  
```

### 5. Calculate the error of the model

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions)) #หาค่าความต่างระหว่าง y_test prediction อาจใช้เปรียบเทียบระหว่างโมเดล
```

```python
print('MSE:', metrics.mean_squared_error(y_test, predictions))
```

```python
import numpy as np
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
