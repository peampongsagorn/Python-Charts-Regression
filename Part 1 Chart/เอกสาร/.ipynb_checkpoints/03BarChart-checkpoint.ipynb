# Bar chart

### 1. Simple bar plot

```python
import pandas as pd
df = pd.read_excel('data.xlsx',sheet_name='drink')
```

```python
brand = pd.unique(df['Brand Purchased'])
y = [0] * len(brand)
for i in range(len(brand)):
    y[i] = df['Brand Purchased'][df['Brand Purchased']==brand[i]].count()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
```

```python
axes = fig.add_axes([0,0,1,1])
```

```python
axes.bar(brand, y, width=0.5)
```

```python
axes.set_xlabel('Brand')
axes.set_ylabel('Number of people purchasing')
axes.set_title('Market survey')
```

```python
fig
```

### 2. Multiple bar plot

```python
import pandas as pd
df = pd.read_excel('data.xlsx',sheet_name='bts')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
```

```python
axes = fig.add_axes([0,0,1,1])
```

```python
df_2016 = df[ (df['date']<'2017-01-01') & (df['date']>'2015-12-31')]
df_2016['month'] = df_2016['date'].dt.month_name()
```

```python
axes.bar(df_2016['month'], df_2016['Average daily ridership'],width=0.5)
```

```python
fig.set_size_inches(11,4)
```

```python
df_2015 = df[ (df['date']<'2016-01-01') & (df['date']>'2014-12-31')]
df_2015['month'] = df_2015['date'].dt.month_name()
```

```python
axes.bar(df_2015['month'], df_2015['Average daily ridership'],width=0.5)
```

```python
fig #
```

#### 

```python
fig = plt.figure()
```

```python
axes = fig.add_axes([0,0,1,1])
```

```python
import numpy as np
m = df_2015['month']
x = np.arange(len(m))
```

```python
width=0.3
ax2015 = axes.bar(x-width/2, df_2015['Average daily ridership'],width)
```

```python
ax2016 = axes.bar(x+width/2, df_2016['Average daily ridership'],width)
```

```python
axes.set_xlabel('Month')
axes.set_ylabel('Ridership')
axes.set_title('BTS')
```

```python
ax2015.set_label('2015')
ax2016.set_label('2016')
axes.legend()
```

```python
axes.set_xticks(x)
axes.set_xticklabels(m)
```

```python
fig.set_size_inches(11,3)
fig.set_dpi(65)
fig
```

#### plot three set of data

```python
import pandas as pd
df = pd.read_excel('data.xlsx',sheet_name='bts')
```

```python
df_2016 = df[ (df['date']<'2017-01-01') & (df['date']>'2015-12-31')]
df_2016['month'] = df_2016['date'].dt.month_name()
```

```python
df_2015 = df[ (df['date']<'2016-01-01') & (df['date']>'2014-12-31')]
df_2015['month'] = df_2015['date'].dt.month_name()
```

```python
df_2014 = df[ (df['date']<'2015-01-01') & (df['date']>'2013-12-31')]
df_2014['month'] = df_2014['date'].dt.month_name()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
```

```python
axes = fig.add_axes([0,0,1,1])
```

```python
import numpy as np
m = df_2015['month']
x = np.arange(len(m))
```

```python
width = 0.2
ax2014 = axes.bar(x-width, df_2014['Average daily ridership'],width)
ax2015 = axes.bar(x, df_2015['Average daily ridership'],width)
ax2016 = axes.bar(x+width, df_2016['Average daily ridership'],width)
```

```python
axes.set_xlabel('Month')
axes.set_ylabel('Ridership')
axes.set_title('BTS')
```

```python
ax2014.set_label('2014')
ax2015.set_label('2015')
ax2016.set_label('2016')
axes.legend()
```

```python
axes.set_xticks(x)
axes.set_xticklabels(m)
```

```python
fig.set_size_inches(11,3)
fig.set_dpi(65)
fig
```
